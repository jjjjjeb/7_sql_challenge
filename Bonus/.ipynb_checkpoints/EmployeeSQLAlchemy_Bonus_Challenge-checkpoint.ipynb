```python
# BONUS

%matplotlib notebook

# Reading & Mapping Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# SQL Alchemy
from sqlalchemy import create_engine

# my config file
import config as c
```

```python
engine = create_engine("postgresql://"+ c.sql['user']+":"+c.sql['passwd']+"@localhost/"+c.sql['db'])
conn = engine.connect()
# had to install psycopg2 (??)
```

```python
salary_data = pd.read_sql("SELECT * FROM Salaries", conn)
salary_data.head()
```

```python
title_data = pd.read_sql("SELECT * FROM Titles", conn)
title_data.head()
```

```python
merged_db = pd.merge(title_data, salary_data, how="left", on="emp_no")
merged_db.head()
```

```python
merged_db = merged_db[['emp_no', 'title', 'salary']]
merged_db.head()
```

```python
salary_list = merged_db['salary'].tolist()
```

```python
# Create Histogram of Salary List

n, bins, patches = plt.hist(x=salary_list, bins=25, color='lightgreen',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.55)
plt.xlabel('Salaries')
plt.ylabel('Count of Salary Occurance')
plt.title('Salary Histogram')
plt.text(23, 45, r'$\mu=15, b=3$')
maxfreq = n.max()
plt.ylim(ymax=np.ceil(maxfreq / 8) * 10 if maxfreq % 10 else maxfreq + 10)
plt.tight_layout()
plt.savefig('./Images/emp_salary_histogram.png')
```

```python
title_group = merged_db.groupby('title').mean()
ave_salary_by_title = title_group.drop(columns = 'emp_no')
ave_salary_by_title
```

```python
titles = merged_db['title'].unique()
titles
```

```python
# Create bar chart of salaries by title

colors = ['lightgreen', 'gold','blue','purple','hotpink','aqua','red']
plt.figure(figsize=(15,5))
plt.bar(titles, ave_salary_by_title['salary'], alpha=0.7, color=colors)
plt.grid(alpha=.6)
plt.xlabel('Title')
plt.ylabel('Average Salary')
plt.title('Average Salary by Title')
plt.tight_layout()
plt.savefig('./Images/ave_salary_by_title_bar.png')
plt.show()
```
